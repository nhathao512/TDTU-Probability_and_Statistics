```python
import random
import math

x = [random.randint(1, 6) for i in range(8000)]
X = set(x)
P = [x.count(i)/len(x) for i in X]
FX = sum(P[:3])

EX = 0
for x in X:
  EX = EX + (x * P[x-1])

VarX = 0
for x in X:
  VarX = VarX + (x-EX)*(x-EX)*P[x-1]

SD = math.sqrt(VarX)

print(x)
print(X)
print(P)
print(FX)
print(EX)
print(VarX)
print(SD)
```
